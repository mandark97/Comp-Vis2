```python
import cv2
import os
from glob import glob
import numpy as np
from sklearn.cluster import MiniBatchKMeans
```

```python
def get_imgs(path):
    """
    - returns  a dictionary of all files
    having key => value as  objectname => image path

    - returns total number of files.

    """
    imlist = {}
    for each in glob(path + "*"):
        word = each.split("/")[-1]
        imlist[word] = []
        for imagefile in glob(path+word+"/*"):
            im = cv2.imread(imagefile, 0)
            imlist[word].append(im)

    return imlist
```

```python
imgs = get_imgs("database/")
sift = cv2.xfeatures2d.SIFT_create()
```

```python
desc_list = []
for word, img_list in imgs.items():
    for img in img_list:
        kp, desc = sift.detectAndCompute(img, None)

        desc_list.append(desc)
```

```python
def format(l):
    vStack = np.array(l[0])
    for remaining in l[1:]:
        vStack = np.vstack((vStack, remaining))
    return vStack
```

```python
formated_desc = format(desc_list)
```

```python
N_CLUSTER = 10000
```

```python
kmeans = MiniBatchKMeans(n_clusters = N_CLUSTER, verbose=10)
```

```python
kmeans.fit(formated_desc)
```

```python
from joblib import dump, load
```

```python
dump(kmeans, f'kmeans{N_CLUSTER}.joblib')
```

```python
hists = []
for desc in desc_list:
    hists.append(np.histogram(kmeans.predict(desc), bins=N_CLUSTER, density=True)[0])
```
